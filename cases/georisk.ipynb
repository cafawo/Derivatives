## Case: Geopolitical Risk

Its **28 February 2026** (the beginning if the 2026 Iran war) and you are in charge of risk management at one of the following companies:

1. Lufthansa
2. BASF
3. Hapag-Lloyd
4. Adidas
5. Volkswagen
6. Heidelberg Materials

(Your company will be assigned in class.)

### Part 1: Risk profile

Prepare a short **verbal discussion of no more than 5 minutes** around the following questions. Assume that your company is known; therefore, keep the introduction concise and focus on the information necessary for the audience to follow your responses. That being said, feel free to provide additional background and insights, for example from company (risk) reporting. Less is more: focus on the main risk drivers. How the discussion is presented, and by whom, is up to you.

#### What were the core risk drivers and hedging strategy before the event?

#### What are the main transmission channels from the shock to your company?

#### Which of these risks are in principle hedgeable with derivatives?

#### Which risks are not really hedgeable with derivatives?

#### If you were the treasurer of the company today (28 February 2026), what would you do?
